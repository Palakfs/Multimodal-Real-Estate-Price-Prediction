```python
cols_to_plot = ['price', 'bedrooms', 'bathrooms', 'sqft_living',
                'floors', 'waterfront', 'view', 'grade',
                'sqft_above', 'sqft_basement', 'sqft_living15']

corr_matrix = df[cols_to_plot].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix,
            annot=True,         # Show numbers
            fmt=".2f",          # Only show 2 decimal places
            cmap='coolwarm',
)
plt.show()
```

```python
df['is_renovated'] = df['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
#feature engineering
```

```python
features = ['bedrooms', 'bathrooms', 'sqft_living','sqft_lot',
                'floors', 'waterfront', 'view','condition','grade',
                'sqft_basement','lat','long','sqft_living15','is_renovated','yr_built']
X = df[features]
y = df['price']
y_log = np.log1p(y)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred_log = model.predict(X_test)

y_pred_original = np.expm1(y_pred_log)
y_test_original = np.expm1(y_test)

rmse = np.sqrt(mean_squared_error(y_test_original, y_pred_original))
r2 = r2_score(y_test_original, y_pred_original)

print(f"Baseline RMSE: ${rmse:,.2f}")
print(f"RÂ² Score: {r2:.4f}")
```

```python
import xgboost as xgb
xgb_model = xgb.XGBRegressor(objective='reg:squarederror',
                             n_estimators=1000,
                             learning_rate=0.05,
                             max_depth=5,
                             random_state=42)
xgb_model.fit(X_train, y_train)
y_pred_log_xgb = xgb_model.predict(X_test)

y_pred_xgb = np.expm1(y_pred_log_xgb)
y_test_original = np.expm1(y_test)

rmse_xgb = np.sqrt(mean_squared_error(y_test_original, y_pred_xgb))
r2_xgb = r2_score(y_test_original, y_pred_xgb)

print(f"XGBoost RMSE: ${rmse_xgb:,.2f}")
print(f"XGBoost RÂ² Score: {r2_xgb:.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred_log_rf = rf_model.predict(X_test)

y_pred_rf = np.expm1(y_pred_log_rf)
y_test_original_rf = np.expm1(y_test)

rmse_rf = np.sqrt(mean_squared_error(y_test_original_rf, y_pred_rf))
r2_rf = r2_score(y_test_original_rf, y_pred_rf)

print(f"Random Forest Regressor RMSE: ${rmse_rf:,.2f}")
print(f"Random Forest Regressor RÂ² Score: {r2_rf:.4f}")
```
